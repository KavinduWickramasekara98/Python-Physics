```python
import math 
sinList =[]
cosList =[]
sinList1 =[]
cosList1 =[]
def calSin(angle):
    return math.sin(angle)
def calCos(angle):
    return math.cos(angle)

angleList=[0,30,45,60,90]
angleListFromPi=[0,(math.pi/6),(math.pi/4),(math.pi/3),(math.pi/2)]
for r in angleList:
    sinList.append(calSin(math.radians(r)))
    cosList.append(calCos(math.radians(r)))
for r in angleListFromPi:
    sinList1.append(calSin(math.degrees(r)))
    cosList1.append(calCos(math.degrees(r)))
    
#a   
print("a) ")
print("sin ", sinList)
print("cos ", cosList)

#b
print("b)")
print("cos 2 ", cosList1)
print("cos2 ", cosList1)
```

```python
import math
import numpy as np 


def maginitudeVector(x,y,z):
    return math.sqrt(x**2+y**2+z**2)


a=[3,-2,1]  #list(map(int, input("Enter x y z: ").split()))
b=[2,-4,-3]   #list(map(int, input("Enter x y z: ").split()))


def magnitude(vector):
    return math.sqrt(sum(pow(element, 2) for element in vector))

def dotproduct(v1, v2):
    return sum((a*b) for a, b in zip(v1, v2))

def length(v):
    return math.sqrt(dotproduct(v, v))

def angle(v1, v2):
    return math.acos(dotproduct(v1, v2) / (length(v1) * length(v2)))


aVec =np.array(a)
bVec =np.array(b)
print("Magnitude of each vector a: ",magnitude(aVec))
print("Magnitude of each vector b: ",magnitude(bVec))
print("a : ",aVec)
print("b : " ,bVec)
print("a+b : " ,np.add(aVec,bVec))
print("a-b : " ,np.subtract(aVec,bVec))
print("a*b : " ,np.multiply(aVec,bVec))
print("a/b : " ,np.divide(aVec,bVec))
print("C=3 : " ,np.multiply(aVec,3))
print("angle : " ,angle(aVec,bVec))
print("Cross product : " ,np.cross(aVec,bVec))
```
